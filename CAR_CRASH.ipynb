## ***Univariate***



```python
injury_counts = df['Injury Type'].value_counts()
injury_counts
```

```python
custom_colors = ['#66b3ff', '#99ff99', 'tomato', 'black']
plt.figure(figsize=(8, 8))
plt.pie(injury_counts, labels=injury_counts.index, autopct='%1.1f%%', startangle=200, colors=custom_colors,labeldistance=None,explode=(0.1,0.1,0.1,0.1),shadow=True)
plt.title('Pie Chart of Injury Types Percentages')
plt.axis('equal')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Collision Type', data=df, color='lightgreen')
plt.title('Distribution of accident')
plt.xlabel('Mode of transportation')
plt.show()
```

```python
plt.hist(df['Hour'], bins=24, edgecolor='black')
plt.title('Distribution of Collisions by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Collisions')
plt.grid(axis='y', alpha=0.75)

plt.show()
```

```python
collision_type_counts = df['Collision Type'].value_counts()
collision_type_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Collisions by Collision Type')
plt.xlabel('Collision Type')
plt.ylabel('Number of Collisions')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
sns.kdeplot(df['Hour'], fill=True, color='coral')
plt.title('Kernel Density Plot of Collisions by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Density')

plt.show()
```

## ***Bivariate***

```python
sns.countplot(data=df,x=df["Weekend?"])
plt.ylabel("Frequency")
plt.title("Number of Accidents")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(x='Primary Factor', y='Hour', data=df)
plt.title('Primary Factor vs. Hour')
plt.xlabel('Primary Factor')
plt.ylabel('Hour')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.countplot(x='Collision Type', hue='Injury Type', data=df)
plt.title('Collision Type vs. Injury Type')
plt.xlabel('Collision Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## **Multivariate**

```python
sns.pairplot(df[['Latitude', 'Longitude', 'Hour']])
plt.suptitle('Pair Plot of Latitude, Longitude, and Hour')
plt.show()
```
